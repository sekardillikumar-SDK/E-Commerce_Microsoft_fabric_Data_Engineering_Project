```python
from pyspark.sql import *
from pyspark.sql.functions import *
```

```python
Customers_Raw = spark.read.format("csv").option("header", True).option("inferSchema", True).load("/Volumes/fabric_lh/default/landing/customers.csv")
orders_Raw = spark.read.format("csv").option("header", True).option("inferSchema", True).load("/Volumes/fabric_lh/default/landing/orders.csv")
payments_Raw = spark.read.format("csv").option("header", True).option("inferSchema", True).load("/Volumes/fabric_lh/default/landing/payments.csv")
support_tickets_Raw = spark.read.format("csv").option("header", True).option("inferSchema", True).load("/Volumes/fabric_lh/default/landing/support_tickets.csv")
web_activities_Raw = spark.read.format("csv").option("header", True).option("inferSchema", True).load("/Volumes/fabric_lh/default/landing/web_activities.csv")
```

```python
Customers_Raw.write.format("delta").mode("overwrite").saveAsTable("fabric_lh.Bronze.customers")
orders_Raw.write.format("delta").mode("overwrite").saveAsTable("fabric_lh.Bronze.orders")
payments_Raw.write.format("delta").mode("overwrite").saveAsTable("fabric_lh.Bronze.payments")
support_tickets_Raw.write.format("delta").mode("overwrite").saveAsTable("fabric_lh.Bronze.support_tickets")
web_activities_Raw.write.format("delta").mode("overwrite").saveAsTable("fabric_lh.Bronze.web_activities")
```
